```python
movies_master = movies_metadata[['budget', 'genres', 'id', 'overview', 'popularity', 'production_companies', 'release_date', 'title']]
movies_master
```

```python
movies_master['release_date'] = pd.to_datetime(movies_master['release_date'], format='%Y-%m-%d')
```

```python
movies_master.info()
```

```python
movies_master.rename(columns={"id": "movie_id"}, inplace=True)
movies_master = movies_master.merge(credits, on='movie_id')
```

```python
movies_master.columns
```

```python
movies_master = movies_master[['genres', 'movie_id', 'overview', 'popularity','production_companies', 'release_date', 'title_x', 'cast', 'crew']]
```

```python
movies_master
```

```python
movies_master.dropna(inplace=True)
```

```python
movies_master
```

```python
import ast
def convert_genres(obj):
  L= []
  for i in ast.literal_eval(obj):
    if i['name']:
      L.append(i['name'])
  return L

def get_cast_from_movie(obj):
  L= []
  count = 0
  for i in ast.literal_eval(obj):
    if i['character'] and count <= 3:
      L.append(i['character'])
      count += 1
    else:
      break

  return L
```

```python
movies_master['genres']=movies_master['genres'].apply(lambda x: convert_genres(x))
movies_master['production_companies']=movies_master['production_companies'].apply(lambda x: convert_genres(x))
movies_master['cast'] = movies_master['cast'].apply(lambda x: get_cast_from_movie(x))
```

```python
movies_master
```

```python
def fetch_director(obj):
  L= []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
  return L
```

```python
movies_master['director'] = movies_master['crew'].apply(lambda x: fetch_director(x))
```

```python
movies_master
```

```python
movies_master.drop('crew', axis=1, inplace=True)
```

```python
movies_master
```

```python
movies_master['overview'] = movies_master['overview'].apply(lambda x: x.split())
```

```python
movies_master
```

```python
movies_master['genres'] = movies_master['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies_master['cast'] = movies_master['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies_master['director'] = movies_master['director'].apply(lambda x: [i.replace(" ", "") for i in x])
movies_master['production_companies'] = movies_master['production_companies'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies_master
```

```python
movies_master['tags'] = movies_master['genres'] + movies_master['overview'] + movies_master['cast'] + movies_master['director'] + movies_master['production_companies']
```

```python
movies_master
```

```python
new_df = movies_master[['movie_id', 'title_x', 'tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x).lower())
```

```python
new_df
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: stem(x))
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=4000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend_movie(movie):
  movie_index = new_df[new_df['title_x'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title_x)
```

```python
recommend_movie('Avatar')
```
